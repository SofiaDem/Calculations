```python
import pandas as pd
import bisect
import numpy as np
import xraylib
import math
numbers = []
intensity_file = open("C:/Beam/Particles.txt", "r")
for i in intensity_file:
    numbers.append(float(i))
print(numbers)
```

```python
I0= np.sum(numbers)
print(I0)
```

```python
xraylib.XRayInit()
sigma= xraylib.CS_Total(74,300)
print(sigma)
```

```python
value = 0.0007
result = []
for x in [1,2,3,4,5]:
    exponent = []
    
    for i in numbers:
        
        m = i/I0*math.exp(-19.25* sigma* x)
        exponent.append(float(m))
        summ = np.sum(exponent)
        
        
    result.append(float(summ))    
    print("Thickness",x, "Transmission",summ)
    #Reference table
    

```
